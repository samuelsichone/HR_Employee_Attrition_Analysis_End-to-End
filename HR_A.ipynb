```python
df.shape
df.columns

df.columns=['age', 'attrition', 'business_travel', 'daily_rate', 'department',
 'distance_from_home', 'education', 'education_field', 'employee_count',
 'employee_number', 'environment_satisfaction', 'gender', 'hourly_rate',
 'job_involvement', 'job_level', 'job_role', 'job_satisfaction',
 'marital_status', 'monthly_income', 'monthly_rate', 'num_companies_worked',
 'over18', 'over_time', 'percent_salary_hike', 'performance_rating',
 'relationship_satisfaction', 'standard_hours', 'stock_option_level',
 'total_working_years', 'training_times_last_year', 'work_life_balance',
 'years_at_company', 'years_in_current_role', 'years_since_last_promotion',
 'years_with_curr_manager']
```

```python
df.head(10)
```

```python
#Found No null values
print("Null values",df.isnull().sum())
print("Information",df.info())
```

```python
df.describe()
```

```python
# Converting Yes/No Columns to 1,0
y_n_Columns=['attrition','over_time']
for col in y_n_Columns:
    df[col]=df[col].map({'Yes':1,'No':0})
```

```python
df.head()
```

```python
#Attrition_Flag For Analysis
df['attrition_flag'] = df['attrition']
df[['attrition', 'attrition_flag']].head()
```

```python
# Age Group For Analysis (Early Career',Mid Career, Senior Career,Senior / Experienced)
def age_group(age):
    if age < 30:
        return 'Early Career'
    elif age < 40:
        return 'Mid Career'
    elif age < 50:
        return 'Senior Career'
    else:
        return 'Senior / Experienced'

df['age_Group'] = df['age'].apply(age_group)

df[['age_Group','age']].head()
```

```python
def salary_band(salary):
    if salary < 4000:
        return 'Low'
    elif salary < 8000:
        return 'Medium'
    else:
        return 'High'

df['salary_band'] = df['monthly_income'].apply(salary_band)
df[['salary_band','monthly_income']].head()
```

```python
# List of columns to clean values
columns_to_clean = ['department', 'education_field', 'job_role']

#  Cleaning values inside these columns (replacing spaces with _ and lowercase)
for col in columns_to_clean:
    df[col] = df[col].str.strip().str.replace(' ', '_').str.lower()
```
